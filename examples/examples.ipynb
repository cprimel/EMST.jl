```julia
iris_features = Iris.features();
(edges, weights, oldfromnew) = EMST.compute_emst(iris_features;leafSize=1);
```

```julia
  scatter(iris_features[1,:],iris_features[2,:], iris_features[3,:], ms=3.0,marker=(stroke(0,:gray)))
  for zr in 1:size(edges,1)
    plot!( iris_features[1,edges[zr,:]] , iris_features[2,edges[zr,:]] , iris_features[3,edges[zr,:]],linecolor = :gray)
  end
  scatter!(legend=false)
```

```julia
for i in 1:size(edges,1)
    println("Edge: $(edges[i,1]) $(edges[i,2]); weight: $(weights[i])")
end
```

```julia
x = Array{Float64,2}(undef, 2, 9)
    
x[1,1], x[2,1] = 0., 0.
x[1,2], x[2,2] = 0., 1.
x[1,3], x[2,3] = 1.0, 2.5
x[1,4], x[2,4] = 0.5, 3.0
x[1,5], x[2,5] = 0.25, 4.
x[1,6], x[2,6] = 4.0, 0.
x[1,7], x[2,7] = 1.0, 3.0
x[1,8], x[2,8] = 6.0, 3.0
x[1,9], x[2,9] = 6.0, 3.0

(x_emst, w_emst, _) = EMST.compute_emst(x;leafSize=1)
```

```julia
scatter(x[1,:],x[2,:], ms=3.0,marker=(stroke(0,:gray)))
for zr in 1:size(x_emst,1)
  plot!( x[1,x_emst[zr,:]] , x[2,x_emst[zr,:]] ,linecolor = :gray)
end
scatter!(legend=false)
```

```julia
for i in 1:size(x_emst,1)
    println("Edge: $(x_emst[i,1]) $(x_emst[i,2]); weight: $(w_emst[i])")
end
```
